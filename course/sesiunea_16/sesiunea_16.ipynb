### Exercițiul 1
**Creează un DataFrame care conține informații despre produse.**  
Coloanele trebuie să fie: `Produs`, `Preț`, `Stoc`.
Populează tabelul cu câteva valori de exemplu.

```python
import pandas as pd

data = {
    "Product": ["Banana", "Apple", "Orange", "Grapes", "Mango", "Pineapple"],
    "Price": [29.99, 7.99, 120.50, 45.00, 35.25, 55.75],
    "Stock": [10, 7, 15, 5, 12, 8],
}

df = pd.DataFrame(data)

df
```

### Exercițiul 2
**Afișează doar produsele care au prețul mai mare de 50.**  
Folosește filtrarea pe coloana `Preț`.

```python
products = df[df["Price"] > 50]

mask = df["Price"] > 50

df[mask]
```

### Exercițiul 3
**Sortează produsele în ordine descrescătoare după preț.**  
Utilizează metoda `sort_values()` din Pandas.

```python
df.sort_values(by="Price", ascending=False)
```

### Exercițiul 4
**Adaugă o coloană nouă numită `TVA`, care reprezintă 19% din prețul fiecărui produs.**

```python
df["TVA"] = [x * 0.19 for x in df["Price"]]

df.sort_values(by="Price", ascending=False)
```

```python
# vectorized
df["TVA"] = df["Price"] * 0.25

df.sort_values(by="Price", ascending=False)
```
