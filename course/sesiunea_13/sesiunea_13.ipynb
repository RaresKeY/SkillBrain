### Exercițiul 1 – Array 2D 5x5 cu valori random între 1 și 100
Creează un array bidimensional (5x5) folosind `NumPy`, care să conțină valori întregi generate aleator între 1 și 100. Afișează rezultatul.

### Exercițiul 2 – Media, maximul și minimul fiecărui rând
Pentru array-ul creat anterior, calculează media, valoarea maximă și valoarea minimă **pentru fiecare rând**.

### Exercițiul 3 – Funcție de normalizare a unui array
Scrie o funcție `normalize(arr)` care primește un array și îl normalizează astfel încât toate valorile să fie în intervalul [0, 1]. Afișează array-ul normalizat.

### Exercițiul 4 – Produs scalar între două array-uri
Creează două array-uri 1D de aceeași lungime și calculează produsul lor scalar folosind funcția `np.dot()`.

### Exercițiul 5 – Verificare simetrie matrice pătratică
Scrie o funcție care verifică dacă o matrice pătratică este **simetrică** (adică egală cu transpusul ei). Testează funcția pe un exemplu concret.
