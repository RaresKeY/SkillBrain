### üìù Exercise 2 ‚Äì 3-Chart Dashboard: Personal Budget

**Goal:** Create a small interactive financial dashboard combining bar, line, and pie charts in a single Plotly figure with subplots.

**Tasks:**
- Prepare 6 months of data: `months`, `income`, `expenses`, and compute `savings` as `income - expenses`.
- Use `make_subplots` to create a 2√ó2 grid of subplots where:
  - Subplot (row=1, col=1): **Bar chart** comparing `Income` vs `Expenses` per month.
  - Subplot (row=1, col=2): **Line chart** (`go.Scatter`) showing `Savings` over the 6 months.
  - Subplot (row=2, col=1..2): **Pie chart** (`go.Pie`) showing one month‚Äôs expense breakdown (e.g. Food, Rent, Transport, Entertainment).
- Add a general title (e.g. *Personal Financial Dashboard ‚Äì 6 Months*) and axis labels where appropriate.
- Make sure the legend is visible and that hovering each trace shows meaningful information.

**Reflection (write in a separate cell or as comments):**
- In which month do you save the most? What insights does the dashboard give about your financial behavior?

```python
# Exercise 2 ‚Äì 3-Chart Dashboard: Personal Budget
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
from pathlib import Path

out_dir = Path("export")
out_dir.mkdir(exist_ok=True)

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
income = [9200, 9800, 10100, 10400, 10800, 11100]
expenses = [7300, 7600, 7900, 8000, 8250, 8400]
savings = [i - e for i, e in zip(income, expenses)]

budget_df = pd.DataFrame({
    "Month": months,
    "Income": income,
    "Expenses": expenses,
    "Savings": savings,
})

expense_breakdown_labels = ["Rent", "Food", "Transport", "Entertainment", "Utilities"]
expense_breakdown_values = [3500, 1800, 900, 700, 1500]  # example for June

fig_dashboard = make_subplots(
    rows=2,
    cols=2,
    specs=[[{"type": "xy"}, {"type": "xy"}], [{"type": "domain"}, {"type": "domain"}]],
    subplot_titles=(
        "Income vs Expenses",
        "Savings Trend",
        "Expense Breakdown (June)",
        "Savings Composition (June)",
    ),
)

fig_dashboard.add_trace(
    go.Bar(name="Income", x=months, y=income, marker_color="#2ca02c"),
    row=1,
    col=1,
)
fig_dashboard.add_trace(
    go.Bar(name="Expenses", x=months, y=expenses, marker_color="#d62728"),
    row=1,
    col=1,
)

fig_dashboard.add_trace(
    go.Scatter(name="Savings", x=months, y=savings, mode="lines+markers", line=dict(color="#1f77b4", width=3)),
    row=1,
    col=2,
)

fig_dashboard.add_trace(
    go.Pie(labels=expense_breakdown_labels, values=expense_breakdown_values, name="Expenses"),
    row=2,
    col=1,
)

fig_dashboard.add_trace(
    go.Pie(labels=["Saved", "Spent"], values=[savings[-1], expenses[-1]], name="June split", hole=0.45),
    row=2,
    col=2,
)

fig_dashboard.update_layout(
    title="Personal Financial Dashboard ‚Äì 6 Months",
    template="plotly_dark",
    height=800,
    barmode="group",
    legend=dict(orientation="h", y=1.08),
)
fig_dashboard.update_xaxes(title_text="Month", row=1, col=1)
fig_dashboard.update_xaxes(title_text="Month", row=1, col=2)
fig_dashboard.update_yaxes(title_text="Amount", row=1, col=1)
fig_dashboard.update_yaxes(title_text="Amount", row=1, col=2)

fig_dashboard.write_html(out_dir / "ex2_budget_dashboard.html", include_plotlyjs="cdn")
fig_dashboard.show()

budget_df
```

### üìù Exercise 3 ‚Äì Box Plot: Team Performance

**Goal:** Compare score distributions for four different teams using an interactive box plot with all underlying data points visible.

**Tasks:**
- Generate around 20‚Äì30 scores for each of 4 teams, e.g. `Team A`, `Team B`, `Team C`, `Team D` (you can use `numpy.random.normal`).
- Build a `DataFrame` with columns `Team` and `Score`.
- Create a box plot using `px.box` (or `go.Box`) with `Team` on the x-axis and `Score` on the y-axis.
- Display all underlying data points on top of the boxes (e.g. `points='all'`).
- Color boxes by `Team`.
- **Bonus:** Add a horizontal target line (e.g. 75 points) with `fig.add_hline(...)`.

**Reflection (write in a separate cell or as comments):**
- Which team appears most consistent (small variation)? Which teams have many outliers?

```python
# Exercise 3 ‚Äì Box Plot: Team Performance
import plotly.express as px
import numpy as np
import pandas as pd
from pathlib import Path

out_dir = Path("export")
out_dir.mkdir(exist_ok=True)

np.random.seed(42)

team_scores = {
    "Team A": np.random.normal(78, 4.5, 28),
    "Team B": np.random.normal(72, 7.0, 28),
    "Team C": np.random.normal(83, 3.0, 28),
    "Team D": np.random.normal(69, 8.5, 28),
}

rows = []
for team, scores in team_scores.items():
    for score in scores:
        rows.append({"Team": team, "Score": float(score)})

df_teams = pd.DataFrame(rows)

fig_box = px.box(
    df_teams,
    x="Team",
    y="Score",
    color="Team",
    points="all",
    title="Team Performance Distribution",
    template="plotly_dark",
    height=520,
)
fig_box.add_hline(y=75, line_dash="dash", line_color="#FFD166", annotation_text="Target = 75")
fig_box.write_html(out_dir / "ex3_team_boxplot.html", include_plotlyjs="cdn")
fig_box.show()

df_teams.groupby("Team")["Score"].describe().round(2)
```

### üìù Exercise 4 ‚Äì Heatmap: 10√ó10 Matrix (and Correlation Matrix Bonus)

**Goal:** Visualize a numeric 10√ó10 matrix as an interactive heatmap and optionally explore a true correlation matrix.

**Tasks:**
- Generate a 10√ó10 matrix with random integer values between 0 and 100 using `numpy`.
- Create a list of labels for both rows and columns (e.g. `Var1`, `Var2`, ‚Ä¶, `Var10`).
- Use `go.Heatmap` to create a heatmap with:
  - `z` = the matrix values.
  - `x` = column labels; `y` = row labels.
  - A meaningful `colorscale` (e.g. `'Viridis'`).
  - `text` and `texttemplate` so the numeric value of each cell is displayed.
- Set a square-ish figure size (e.g. width and height ~700) and suitable titles for axes and the colorbar.
- **Bonus:**
  - Generate a dataset with 10 numeric variables (e.g. 100 observations) and compute the 10√ó10 correlation matrix.
  - Plot the correlation matrix as another heatmap with a diverging colorscale such as `'RdBu'`.

**Reflection (write in a separate cell or as comments):**
- In the correlation heatmap (bonus), which pairs of variables seem strongly positively or negatively correlated?

```python
# Exercise 4 ‚Äì Heatmap: 10x10 Matrix and Correlation Matrix (Bonus)
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
import pandas as pd
from pathlib import Path

out_dir = Path("export")
out_dir.mkdir(exist_ok=True)

np.random.seed(42)
labels = [f"Var{i}" for i in range(1, 11)]
matrix_10 = np.random.randint(0, 101, size=(10, 10))

fig_heat = go.Figure(
    data=go.Heatmap(
        z=matrix_10,
        x=labels,
        y=labels,
        colorscale="Viridis",
        text=matrix_10,
        texttemplate="%{text}",
        colorbar=dict(title="Value"),
    )
)
fig_heat.update_layout(
    title="10√ó10 Random Matrix Heatmap",
    xaxis_title="Variables",
    yaxis_title="Variables",
    template="plotly_dark",
    width=760,
    height=700,
)
fig_heat.write_html(out_dir / "ex4_random_heatmap.html", include_plotlyjs="cdn")
fig_heat.show()

# Bonus: correlation matrix
sample = pd.DataFrame(
    np.random.normal(size=(160, 10)),
    columns=labels,
)
sample["Var3"] = sample["Var1"] * 0.7 + np.random.normal(scale=0.4, size=len(sample))
sample["Var8"] = -sample["Var2"] * 0.65 + np.random.normal(scale=0.45, size=len(sample))

corr = sample.corr(numeric_only=True)

fig_corr = px.imshow(
    corr,
    text_auto=".2f",
    color_continuous_scale="RdBu",
    zmin=-1,
    zmax=1,
    title="Correlation Matrix (Bonus)",
    template="plotly_dark",
    width=760,
    height=700,
)
fig_corr.write_html(out_dir / "ex4_corr_heatmap_bonus.html", include_plotlyjs="cdn")
fig_corr.show()

corr
```

### üöÄ Challenge 1 ‚Äì Interactive Dashboard with Dash

**Goal:** Turn the financial dashboard (Exercise 2) into a small Dash web app where the charts react to user input.

**Tasks:**
- Install Dash if needed (e.g. `pip install dash`).
- Create a `Dash` app with:
  - A dropdown (`dcc.Dropdown`) to select a month.
  - One or more `dcc.Graph` components that update when the dropdown value changes.
- Define a callback (`@app.callback`) that:
  - Receives the selected month as input.
  - Filters or transforms your data accordingly.
  - Returns an updated Plotly figure for display.
- Run the app locally and interact with it in your browser.

You can base the structure on the example given in the handout, but keep the implementation here.

```python
# Challenge 1 ‚Äì Interactive Dashboard with Dash
# Minimal runnable template; run this cell as a script context to launch the app.
from dash import Dash, dcc, html, Input, Output
import plotly.express as px
import pandas as pd

app = Dash(__name__)

app.layout = html.Div([
    html.H3("Dash Mini Dashboard - Monthly Budget"),
    dcc.Dropdown(
        id="month_dd",
        options=[{"label": m, "value": m} for m in budget_df["Month"]],
        value=budget_df["Month"].iloc[-1],
        clearable=False,
        style={"width": "300px"},
    ),
    dcc.Graph(id="month_bar"),
])


@app.callback(Output("month_bar", "figure"), Input("month_dd", "value"))
def update_month_chart(month):
    row = budget_df[budget_df["Month"] == month].iloc[0]
    mini = pd.DataFrame({
        "Metric": ["Income", "Expenses", "Savings"],
        "Value": [row["Income"], row["Expenses"], row["Savings"]],
    })
    fig = px.bar(mini, x="Metric", y="Value", color="Metric", template="plotly_dark", title=f"{month} Snapshot")
    return fig

# To run locally, uncomment:
# if __name__ == "__main__":
#     app.run(debug=True, port=8050)

"Dash template ready. Uncomment app.run(...) to launch."
```

### üìä Challenge 2 ‚Äì Animated Chart

**Goal:** Build an animated scatter (or other chart) that shows how your data evolves over a discrete time dimension.

**Tasks:**
- Create or reuse a time-series style dataset containing:
  - A time variable (e.g. `Year`, `Month`, or `Frame`).
  - At least two numeric variables to plot (e.g. `Age`, `Salary`, `Experience`).
  - One or more categorical variables for `color` or `facet` if useful.
- Use `px.scatter` (or another suitable `px.*` function) with the `animation_frame` argument set to your time variable.
- Optionally vary marker size (`size`) and color (`color`) for more information per frame.
- Play the animation and experiment with the controls (play/pause, slider, etc.).

```python
# Challenge 2 ‚Äì Animated Chart
import plotly.express as px
import pandas as pd
import numpy as np
from pathlib import Path

out_dir = Path("export")
out_dir.mkdir(exist_ok=True)

np.random.seed(42)
frames = []
for year in range(2019, 2025):
    for field in ["Data", "Web", "DevOps", "Management"]:
        base_salary = {"Data": 6500, "Web": 5600, "DevOps": 7100, "Management": 9800}[field]
        avg_exp = {"Data": 5, "Web": 4, "DevOps": 6, "Management": 10}[field]
        frames.append({
            "Year": year,
            "Field": field,
            "Avg_Experience": avg_exp + np.random.normal(0, 0.6),
            "Avg_Salary": base_salary + (year - 2019) * np.random.randint(250, 550) + np.random.normal(0, 250),
            "Team_Size": np.random.randint(8, 35),
        })

anim_df = pd.DataFrame(frames)

fig_anim = px.scatter(
    anim_df,
    x="Avg_Experience",
    y="Avg_Salary",
    size="Team_Size",
    color="Field",
    animation_frame="Year",
    range_x=[2.5, 12],
    range_y=[4500, 17000],
    template="plotly_dark",
    title="Animated Evolution: Experience vs Salary",
)
fig_anim.write_html(out_dir / "challenge2_animated_scatter.html", include_plotlyjs="cdn")
fig_anim.show()
```

### üó∫Ô∏è Challenge 3 ‚Äì Geographic Map

**Goal:** Display data on a geographic map (e.g. by region, county, or country) using Plotly.

**Tasks:**
- Build a dataset where each row corresponds to a geographic unit (e.g. country, state, or county) and at least one numeric metric (e.g. average salary, population, sales).
- Use a suitable Plotly mapping function (e.g. `px.choropleth` or `px.choropleth_mapbox`).
- Map the locations appropriately (e.g. ISO country codes, FIPS codes, or region names, depending on the dataset).
- Use `color` to represent the numeric variable and add a meaningful colorbar title.
- Add an informative figure title and try out the map interactivity (zoom, pan, hover).

```python
# Challenge 3 ‚Äì Geographic Map
import plotly.express as px
import pandas as pd
from pathlib import Path

out_dir = Path("export")
out_dir.mkdir(exist_ok=True)

map_df = pd.DataFrame({
    "iso_alpha": ["ROU", "DEU", "FRA", "ESP", "ITA", "NLD", "POL", "SWE", "USA", "CAN"],
    "country": ["Romania", "Germany", "France", "Spain", "Italy", "Netherlands", "Poland", "Sweden", "USA", "Canada"],
    "avg_salary_index": [62, 88, 84, 77, 79, 91, 70, 93, 100, 95],
})

fig_map = px.choropleth(
    map_df,
    locations="iso_alpha",
    color="avg_salary_index",
    hover_name="country",
    color_continuous_scale="Plasma",
    title="Average Salary Index by Country",
    labels={"avg_salary_index": "Salary Index"},
    template="plotly_dark",
)
fig_map.update_layout(height=550)
fig_map.write_html(out_dir / "challenge3_geo_map.html", include_plotlyjs="cdn")
fig_map.show()
```
