```python
# One-time build reverse search index, without using classes
reverse_music_index = dict()

# 2-3 depth nested dict
for key_0, item_0 in music.items():
    for key_1, item_1 in item_0.items():
        if not isinstance(item_1, (list, set, dict)):
            reverse_music_index.setdefault(item_1, []).append(key_0)
        else:
            for l_item in item_1:
                reverse_music_index.setdefault(l_item, []).append(key_0)

# check
print(len(reverse_music_index))
for key, item in reverse_music_index.items():
    print(key, '', item)
```

## Sorting

```python
# sorting for rating/duration

ratings = []
durations = []

for key, item in music.items():
    ratings.append((item["title"], item["rating"]))
    durations.append((item["title"], item["duration"]))

ratings.sort(key=lambda x: x[1], reverse=True)
durations.sort(key=lambda x: x[1])

print(ratings)
print(durations)
```

# Moody AI Select

```python
# Local AI stuff

import os
import requests
import json
import re

# pull configuration from environment variables with fallbacks
BASE_URL = os.environ.get("LLM_BASE_URL", "http://localhost:8080/v1")
API_KEY  = os.environ.get("LLM_API_KEY", "dummy")
MODEL    = os.environ.get("LLM_MODEL", "gemma-3-12b-it-qat-Q4_K_M.gguf")
TIMEOUT  = 120

# endpoint request (json, similar to dict struct)
def get_song_for_mood(mood: str, *, debug: bool=False) -> str:
    music_list_basic = [{"title": values["title"], "genre": values["genre"], "mood_tags": values["mood_tags"]} for values in music.values()]
    if debug: print(music_list_basic)
    payload = {
            "model": MODEL,
            "temperature": 0.2,
            "messages": [
                {
                    "role": "system",
                    "content": (
                        "Select exactly one candidate title for the given mood. "
                        "Respond with a SINGLE JSON object only, no prose. "
                        "Return EXACT title, do not reformat, be especially careful of quotes"
                        "Keys: title, why. "
                        "Constraints: "
                        "1) title MUST be EXACTLY one of the provided titles, "
                        "2) why = one flavorful, user-exposed sentence explaining the choice."
                    ),
                },
                {
                    "role": "user",
                    "content": (
                        f"Mood: {mood}\n\n"
                        "Candidates (JSON array):\n"
                        f"{music_list_basic}"
                    ),
                },
            ],
        }
    
    r = requests.post(
        f"{BASE_URL}/chat/completions",
        headers={"Authorization": f"Bearer {API_KEY}"},
        json=payload,
        timeout=TIMEOUT,
    )
    r.raise_for_status()

    content = r.json()["choices"][0]["message"]["content"].strip()

    # Robust JSON parse (strip any accidental wrapping text)
    m = re.search(r"\{.*\}", content, flags=re.S)
    data = json.loads(m.group(0)) if m else json.loads(content)
    if debug: print("DEBUG:", data)
    titles = [song["title"] for song in music.values()]

    # Optional safety: ensure title is one of the provided titles
    if data.get("title") not in titles:
        # fall back to first candidate if model drifted
        data = {"title": titles[0], "why": "Model returned unknown title; defaulted to first candidate."}

    return {"title": data.get("title", titles[0]), "why": data.get("why", "No reason provided.")}


def display_helper(entry): # quick helper since I keep writing the same stuff
    print(str(entry["title"]).upper())
    for key, item in music[entry["title"]].items():
        print(" ", str(key).capitalize().ljust(9), ": ", str(item).capitalize(), sep="")
```
